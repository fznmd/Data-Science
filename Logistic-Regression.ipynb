# Logistic Regression

```
data = {
    "col" : [1, 2, 3, 4],
    "col1" : [2, 3, 4 ,5],
    "col2" : [0, 1, 0, 1]
}
```

```
data = pd.DataFrame(data=data)
```

```
data
```

```
data.shape
```

```
w = np.random.randn(2, 4)
```

```
w[0]
```

```
data['col'] * w[0]
```

```
y = w[0] * data['col'] + w[1] * data['col1'] + 1
```

```
y
```

```
y_pred = 1/(1+np.exp(-y))
y_pred
```

```
final_y_pred = [1 if final_y >= 0.5 else 0 for final_y in y_pred]
```

```
final_y_pred
```

```
data['col2'], final_y_pred, 
```

```
from sklearn.metrics import accuracy_score

accuracy_score(data["col2"], final_y_pred)
```
